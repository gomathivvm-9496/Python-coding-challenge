```python
 #While Loop, else, break, continue
#Project Title: Shopping Cart System

item_count=0
subtotal=0

while True:
  item_price=float(input("Enter the item price:"))
  if item_price==0:
    break

  if item_price<0:
    print("Invalid amount!Enter a positive value")
    continue
  item_count+=1
  subtotal+=item_price


if subtotal>=1000:
  Discount_rate=0.20
elif subtotal>=500:
  Discount_rate=0.10
elif subtotal>=200:
  Discount_rate=0.05
else:
  Discount_rate=0.0

discount_amount=subtotal*Discount_rate
total_amount=subtotal-discount_amount


print("-----Bill summery----")
print("item count:",item_count)
print("Total price of items:",subtotal)
print("discount rate:",Discount_rate)
print("discount amount:",discount_amount)
print("Final amount to pay:",total_amount)
```

```python
#Built-in Functions and User-Defined Functions
#Project Title: Simple Calculator App
def add(x,y):
  return x+y
def subtract(x,y):
  return abs(x-y)
def multiply(x,y):
  return pow(x,1)*pow(y,1)
def divide(x,y):
  if y==0:
    return None
  return round(x/y,2)

num1=float(input("Enter the first number:"))
num2=float(input("Enter the second number:"))
operation=input("Enter the operation to do (+,-,*,/):")

if operation=="+":
  result=add(num1,num2)
  print("Result:",result)

elif operation=="-":
  result=subtract(num1,num2)
  print("Result:",result)

elif operation=="*":
  result=multiply(num1,num2)
  print("Result:",result)

elif operation=="/":
  result=add(num1,num2)
  if result==None:
    print("division by zero is not possible")
  else:
    print("Result",result)

else:
  print("Invalid operation selected")
```

```python
#Function Arguments, Local and Global Namespace, Introduction to OOP
 #Project Title: Bank Account Manager

transaction_count=0

class BankAccount:
  def __init__(self,account_holder,balance=0):
    self.account_holder=account_holder
    self.balance=balance
  def deposit(self,amount):
    global transaction_count
    self.balance+=amount
    transaction_count+=1
    print(f"₹{amount} deposited successfully.New balance is {self.balance}")
  def withdraw(self,amount):
    global transaction_count
    if amount>self.balance:
      print("Withdraw denied.Insufficient balance")
    else:
      self.balance-=amount
      transaction_count+=1
      print(f"₹{amount} withdraw successfully.New balance is {self.balance}")
  def display_balance(self):

    print(f"Account holder:{self.account_holder}")
    print(f"Current balance:{self.balance}")

account1=BankAccount("Radhi",600)

#positional arguments
account1.deposit(500)
account1.withdraw(300)

#keyword arguments
account1.deposit(amount=200)
account1.withdraw(amount=1500)


account1.display_balance()
print("Transaction count:",transaction_count)








```
